```python
import scipy.stats

def CalcRSqaured(observed, estimated):
    """Calculate the r^2 from a series of observed and estimated target values
    inputs:
    Observed: Series of actual observed values
    estimated: Series of predicted values"""
    
    r, p = scipy.stats.pearsonr(observed, estimated)
    R2 = r **2
    
    return R2

```

```python
from math import sqrt

def CalcRMSE(observed, estimated):
    """Calculate Root Mean Square Error between a series of observed and estimated values
    inputs:
    Observed: Series of actual observed values
    estimated: Series of predicted values"""
    
    res = (observed -estimated)**2
    RMSE = round(sqrt(res.mean()), 3)
    
    return RMSE
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd

data = pd.read_csv('/Users/liuzhengyong/Desktop/CASA Study/Urben Simulation/Assessment/london_flows.csv')
data.replace([np.inf, -np.inf], np.nan, inplace=True)
data = data.dropna(subset=["jobs", "distance"])
data = data[ (data["jobs"] > 0) & (data["distance"] > 0)]

constant = 0.1
data["jobs"] += constant
data["distance"] += constant
x_variables = ["jobs", "distance"]
log_x_vars = []
for x in x_variables:
    log_col_name = f"log_{x}"
    data[log_col_name] = np.log(data[x])
    log_x_vars.append(log_col_name)

formula = 'flows ~ ' + ' + '.join(log_x_vars)
try:
    uncosim = smf.glm(formula=formula, data=data, family=sm.families.Poisson()).fit()
    print(uncosim.summary())
except Exception as e:
    print("Error in model fitting:", e)
```

```python
K = uncosim.params[0]
alpha = uncosim.params[1]
beta = -uncosim.params[2]
data["unconstrainedEst2"] = np.exp(K + alpha * data["log_jobs"] - (beta) * data["log_distance"])

predictions = uncosim.get_prediction()
predictions_summary_frame = predictions.summary_frame()
data["fitted"] = predictions_summary_frame["mean"]
```
